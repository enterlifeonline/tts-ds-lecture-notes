If you scan over a dozen job post for a data scientist it will become clear to you that there is no standard definement of the data scientist role. This definition can be further confused by the fact that there are other roles sometimes thought of as the same, but are often quite different. Some of these include data analyst, data engineer, machine learning engineer and so on. We will discuss more on that in a bit. 

I define data science as the art of uncovering the insights and trends in data in order to deliver [business] value. 

I define a data scientist in the corporate world as someone who finds solutions to known and unknown problems by analyzing data. In addition a data scientist is a story teller who able to effectively communicate their findings in digestable pieces in order for others to quickly grasp what they are saying. 

While the tool box may vary from data scientist to data scientest, the best data scientist will have an expertise in the following 4 areas: 

1. Business Knowledge 
2. Mathmatics ( Specifically statistics ) 
3. Computer Science ( Coding skills like sql, python, r )
4. Communication 

Let's be honest, finding someone who has mastered all four of these is much harder than you think. Here is one of my favorite data scientist Venn diagrams created by Stephan Kolassa.As this diagram shows, Stephan Kolassa labels The Perfect Data Scientist as the individual who is equally strong in business, programming, statistics, and communication.
