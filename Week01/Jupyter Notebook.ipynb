## What is Jupyter Notebook? 

I think [Jupyter Notebook](https://jupyter-notebook.readthedocs.io/en/stable/examples/Notebook/What%20is%20the%20Jupyter%20Notebook.html) explains it best, and please reference this website as you get to know jupyter notebook. 

#### Writing text instead of code

You need to change you cell selection from code to markdown in order to write text

#### Formatting Text 
1. To bold put a double underscore __ before and after the word/sentence or use a ** before and after the word/sentenc
2. For italics use a single underscore _ or a single aestrick * before and after the word/sentence
3. For a forced line break put < br > ( *no spaces* )
4. To change the heading size add a pound size before the first word. The more pound signs the smaller the heading.
5. Use the format in the image in order to change the font 
6. Use 3 aestricks in order to create a horizontal seperator
7. Use the greater than sign to indent 
8. To embed a link, put the name you want to show in brackets and immediately after with no space put the link contained in parentheses.
